```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.histplot(data=df2, x='Credit_Amount', binwidth=1000000)
plt.show()
```

#🧩 1. Grafik nimani anglatadi

###Gorizontal o‘q (x) — kredit miqdori (Credit_Amount), ya’ni qancha pul berilgan.

###Vertikal o‘q (y) — shu miqdorga to‘g‘ri keladigan mijozlar soni (Count).


###Shuning uchun bu grafikda har bir ustun — muayyan oraliqdagi kredit summasini olgan mijozlar sonini ko‘rsatadi.


#📊 2. Grafikdagi holat

###Ustunlar deyarli barqaror (tekis) va keng oralig‘da tarqalgan.
###Bu shuni anglatadi:

> Kredit miqdorlari aniq bitta qiymat atrofida jamlanmagan — turli mijozlar turli summalarda kredit olishgan.

###Ya’ni:

###Bank bir xil hajmdagi kreditlarni emas, balki har xil miqdorlarda (kichik, o‘rta, katta) kreditlar bergan.

###Bu diversifikatsiya degani — risklar turlicha taqsimlangan.


###Bu tekis taqsimot quyidagi holatlardan birini anglatadi:

###Ma’lumotlar random tarzda (tasodifiy) yaratilganligi(mock data’da bu tabiiy).

###Haqiqiy bank ma’lumotlarida histogram odatda chap tomonda balandroq bo‘ladi, ya’ni:

###ko‘pchilik mijozlar kichik kredit olgan (masalan 5–20 mln),

###kam sonli mijozlar katta kredit olgan (100 mln+).

```python
sns.histplot(data=df2, x='Credit_Amount', log_scale=True, bins=50)
plt.show()
```
